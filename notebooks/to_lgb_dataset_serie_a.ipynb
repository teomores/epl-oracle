```python
import lightgbm as lgb

from sklearn.model_selection import train_test_split

X = final_df.drop(['label'],axis=1)
y = final_df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15,shuffle=False)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, shuffle=False)
```

```python
from skopt.space import Real, Integer
from skopt.utils import use_named_args
space  = [Integer(1, 15, name='max_depth'),
          Real(10**-4, 10**0, "log-uniform", name='learning_rate'),
          Integer(50, 1000, name='n_estimators'),
          Real(0,0.01, name='reg_alpha'),
          Real(0,0.01, name='reg_lambda'),
        Integer(3,50,name='num_leaves'),
          Integer(5,30, name=' min_child_samples')
]

@use_named_args(space)
def objective(**params):
    model = lgb.LGBMClassifier(**params)
    model.fit(X_train, y_train)
    predicted_y = model.predict(X_test)
    correct = 0
    for ind,true,predicted in zip(y_test.index, y_test, predicted_y):
        if true == predicted:
            correct+=1
    return -(correct*100 / y_test.shape[0])
```

```python
from skopt import forest_minimize
import numpy as np
res_gp = forest_minimize(objective, space, n_calls=100, random_state=42, verbose=False)

print("Best score=%.4f" % res_gp.fun)
print("""Best parameters:
- max_depth=%d
- learning_rate=%.6f
- n_estimators=%d
- reg_alpha=%d
- reg_lambda=%d
- num_leaverS=%d
-  min_child_samples=%d""" 
      % (res_gp.x[0], res_gp.x[1], 
                            res_gp.x[2], res_gp.x[3], 
                            res_gp.x[4],res_gp.x[5], res_gp.x[6]))
```

```python
model = lgb.LGBMClassifier(max_depth=1,learning_rate=0.003129, n_estimators=975, num_leaves=6,min_child_samples=14)
model.fit(X_train, y_train)
predicted_y = model.predict(X_val)
correct = 0
for ind,true,predicted in zip(y_val.index, y_val, predicted_y):
    if true == predicted:
        correct+=1
correct*100 / y_val.shape[0]
```
