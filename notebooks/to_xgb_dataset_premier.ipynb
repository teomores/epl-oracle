```python
tmp = pd.merge(user_pref_df, class_df.add_prefix('home_'), how='left', left_on=['league_year','match_round','home_team'], right_on=['home_league_year','home_match_round','home_team']).fillna(0).drop(['home_league_year','home_match_round'],axis=1)
tmp = pd.merge(tmp, class_df.add_prefix('away_'), how='left', left_on=['league_year','match_round','away_team'], right_on=['away_league_year','away_match_round','away_team']).fillna(0).drop(['away_league_year','away_match_round'],axis=1)
tmp[tmp.columns[7:]] = tmp[tmp.columns[7:]].astype(int)
tmp.head()
```

```python
tmp_df = pd.merge(df, tmp, how='left',on=['league_year', 'match_round','home_team','away_team'])
tmp_df.head()
```

```python
sorted(tmp_df.columns)
```

```python
final_df = tmp_df.drop(['match_date','home_team','away_team','home_goals','away_goals'], axis=1)
final_df = final_df.sort_values(by = ['league_year', 'match_round']).reset_index(drop=True)
final_df = final_df.fillna(0)
final_df.head()
```

```python
sorted(final_df.columns)
```

```python
print(f"# pareggi : {list(final_df.label).count(0)} ({list(final_df.label).count(0)*100/final_df.shape[0]:.2f})")
print(f"# vittorie casa : {list(final_df.label).count(1)} ({list(final_df.label).count(1)*100/final_df.shape[0]:.2f})")
print(f"# vittorie fuori casa : {list(final_df.label).count(2)} ({list(final_df.label).count(2)*100/final_df.shape[0]:.2f})")
```

```python
import lightgbm as lgb

from sklearn.model_selection import train_test_split

X = final_df.drop(['label'],axis=1)
y = final_df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15,shuffle=False)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, shuffle=False)
```

```python
from skopt.space import Real, Integer
from skopt.utils import use_named_args
space  = [Integer(1, 20, name='max_depth'),
          Real(10**-4, 10**0, "log-uniform", name='learning_rate'),
          Integer(50, 1500, name='n_estimators'),
        Integer(3,50,name='num_leaves'),
          Integer(5,30, name=' min_child_samples')
]

@use_named_args(space)
def objective(**params):
    model = lgb.LGBMClassifier(**params)
    model.fit(X_train, y_train)
    predicted_y = model.predict(X_test)
    correct = 0
    for ind,true,predicted in zip(y_test.index, y_test, predicted_y):
        if true == predicted:
            correct+=1
    return -(correct*100 / y_test.shape[0])
```

```python
from skopt import forest_minimize
import numpy as np
res_gp = forest_minimize(objective, space, n_calls=10, random_state=42, verbose=False)

print("Best score=%.4f" % res_gp.fun)
print("""Best parameters:
- max_depth=%d
- learning_rate=%.6f
- n_estimators=%d
- num_leaver=%d
-  min_child_samples=%d""" 
      % (res_gp.x[0], res_gp.x[1], 
                            res_gp.x[2], res_gp.x[3], 
                            res_gp.x[4]))
```

```python
model = lgb.LGBMClassifier(max_depth=4,learning_rate=0.001287, n_estimators=868, num_leaves=34,min_child_samples=29)
model.fit(X_train, y_train)
predicted_y = model.predict(X_val)
correct = 0
for ind,true,predicted in zip(y_val.index, y_val, predicted_y):
    if true == predicted:
        correct+=1
correct*100 / y_val.shape[0]
```

```python
# baseline che usa solo le quote
result = []
for p1,pX,p2 in zip(X_val['10bet_1'], X_val['10bet_X'], X_val['10bet_2']):
    if min([p1,pX,p2]) == p1: result.append(1)
    elif min([p1,pX,p2]) == pX: result.append(0)
    else: result.append(2)

correct = 0
for ind,true,predicted in zip(y_val.index, y_val, result):
    if true == predicted:
        correct+=1
correct*100 / y_val.shape[0]
```
