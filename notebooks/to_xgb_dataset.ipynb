```python
import pandas as pd
df = pd.read_csv('../dataset/serie_a/transformed/transformed_merged_serie_a.csv')
df.head()
```

```python
label = []
for hg,ag in zip(df.home_goals, df.away_goals):
    if hg == ag:
        label.append(0)
    elif hg > ag:
        label.append(1)
    else:
        label.append(2)
df['label'] = label
df.head(20)
```

```python
for i in range(df.shape[0]):
     df.at[i, 'match_hour'] = int(df.at[i, 'match_hour'].split(':')[0])
```

```python
match_month, match_day, match_year = [], [], []
for date in df.match_date:
    match_day.append(int(date.split('.')[0]))
    match_month.append(int(date.split('.')[1]))
    match_year.append(int(date.split('.')[2]))
```

```python
df['match_day'] = match_day
df['match_month'] = match_month
df['match_year'] = match_year
df.match_hour = df.match_hour.astype(int)
```

```python
df.sort_values(by=['match_year','match_round','home_team','away_team']).head(19)
```

```python
user_pref_df.sort_values(by=['match_year','match_round','home_team','away_team']).head(19)
```

```python
tmp_df = pd.merge(df, user_pref_df, on=['match_year', 'match_round','home_team','away_team'])
tmp_df.head()
```

```python
final_df = tmp_df.drop(['match_date','home_team','away_team','home_goals','away_goals'], axis=1)
final_df = final_df.sort_values(by = ['match_year', 'match_month','match_day','match_hour','match_round']).reset_index(drop=True)
final_df.head()
```

```python
print(f"# pareggi : {list(final_df.label).count(0)} ({list(final_df.label).count(0)*100/final_df.shape[0]:.2f})")
print(f"# vittorie casa : {list(final_df.label).count(1)} ({list(final_df.label).count(1)*100/final_df.shape[0]:.2f})")
print(f"# vittorie fuori casa : {list(final_df.label).count(2)} ({list(final_df.label).count(2)*100/final_df.shape[0]:.2f})")
```

```python
import lightgbm as lgb

from sklearn.model_selection import train_test_split

X = final_df.drop(['label'],axis=1)
y = final_df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,shuffle=False)

model = lgb.LGBMClassifier(learning_rate=0.05,
                          num_leaves=3, n_estimators=70, reg_lambda=0.1
                          )
model.fit(X_train, y_train)

# mio modello
predicted_y = model.predict(X_test)
correct = 0
for ind,true,predicted in zip(y_test.index, y_test, predicted_y):
    if true == predicted:
        correct+=1
correct*100 / y_test.shape[0]
```

```python
# baseline che usa solo le quote
result = []
for p1,pX,p2 in zip(X_test['188bet_1'], X_test['188bet_X'], X_test['188bet_2']):
    if min([p1,pX,p2]) == p1: result.append(1)
    elif min([p1,pX,p2]) == pX: result.append(0)
    else: result.append(2)

correct = 0
for ind,true,predicted in zip(y_test.index, y_test, result):
    if true == predicted:
        correct+=1
correct*100 / y_test.shape[0]
```
